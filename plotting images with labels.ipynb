```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import cv2
import os
```

```python
def PlotImages(ix_start,num_img,rows,cols,images,labels,color=True):
    plt.figure(figsize=(10,10))
    plt.subplots_adjust(wspace=0.5,hspace=0.5)
    for i in range(num_img):
        plt.subplot(rows,cols,i+1)
        plt.axis('off')
        if color:
            plt.imshow(images[ix_start+i])
        else: 
            plt.imshow(np.squeeze(images[ix_start+i]),cmap='Greys')
        #modified to show correct labels
        plt.title(labels[ix_start+i])
```

```python
def load_images(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images
```

```python
images = load_images('./Data/dataset-master/JPEGImages/')
```

```python
df = pd.read_csv('./Data/dataset-master/labels.csv')
df.head()
```

```python
# get rid of the first column
df = df.drop(df.columns[0],axis=1)
df.head()
```

```python
labels = df.Category
```

```python
PlotImages(0,25,5,5,images,labels)
```
